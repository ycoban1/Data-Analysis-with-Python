# NumPy Indexing and Selection

In this lecture we will discuss how to select elements or groups of elements from an array.

```python
import numpy as np
```

```python
arr=np.arange(0,11)
arr
```

## Bracket Indexing and Selection
The simplest way to pick one or some elements of an array looks very similar to python lists:

```python
arr[8] #8. indeksi getirir.
```

```python
arr[-3] #sondan 3. getirir.
```

```python
arr[2:5] #2 ile 5. indek arasını getirir.
```

```python
arr[:5] #0 ile 5. indeks arasını getirir.
```

```python
arr[1::2] #tek sayıları getirir.
```

```python
arr[::2] #çift sayıları getirir.
```

## Indexing a 2D array (matrices)

<p>The general format is <b>arr_2d[row][col]</b> or <b>arr_2d[row,col]</b>. I recommend usually using the comma notation for clarity.</p>

```python
arr_2d=np.array([[5,10,15],[20,25,30],[35,40,45]])
arr_2d
```

```python
arr_2d[1] #1. rav
```

```python
arr_2d[:,1] #1. col
```

```python
arr_2d[1,1] #1. row 1. col olanı al
```

```python
arr_2d[:3,1:] #2. row a kadar. 1 col dan itibaren
```

```python
arr_2d[1:,:2] #1.row dan itibaren, 2. col a kadar
```

```python
arr_2d[2,1]=99 #kalıcı değişiklik yaparak (2,1) değerini 99 yapar
```

```python
arr_2d[:,:] #arr_2d ile aynı karşılık
```

### Fancy Indexing

Fancy indexing allows you to select entire rows or columns out of order,to show this, let's quickly build out a numpy array:

```python
v=np.arange(0,30,3) #0 dan 30 a kadar 3 er aralıklarla
v
```

```python
v[[3,5]] # 2 boyutlu olduğu için iki tane köşeli parantez olmalı. Tek köşeli olsaydı içindeki de tek bir rakam olmalıydı
```

```python
arr2d=np.zeros((10,10),dtype=int)
arr2d
```

```python
for i in range(len(arr2d)):
  arr2d[i]=i
arr2d
```

```python
arr2d[[2,4,6]] #2,4 ve 6. satırları alır
```

<h3>any_array[[row indices], [column indices]]</h3>

```python
jj=np.arange(1,17).reshape(4,4) #1-17 arasındaki sayılardan 4x4 oluşturur.
jj
```

```python
jj[[1,2],[0,3]] #1. row 0.col, 2.raw,3.col alır. jj[(1,2),(0,3)] ile aynı olur
```

```python
jj[2,[2,3]] #2.row,2.col ve 2.row, 3.col. jj[2,2:4] ile aynı olur
```

```python
jj[[2,3],2]
```

```python
ii=np.arange(0,12).reshape(2,6)
ii
```

### Using ***basic index*** and ***fancy index*** together

```python
jj[1:3,[1,2]] #1-3 arası row, 1. ve 3. col
```

```python
jj[[2,3],1] #2. ve 3. row, 1. col
```

### Using ***basic slicing*** and ***fancy index*** together

```python
jj
```

```python
jj[1:3, [1,2]] #1-3 row, 1. ve 2. col
```

```python
jj[1:3, [1,3]] #1-3 row, 1. ve 3. col
```

# Selection on a Condition

Let's briefly go over how to use brackets for selection based off of comparison operators.

```python
arr = np.arange(1,11)
arr
```

```python
arr>4 #4'den büyük olanlara True dedi
```

```python
arr[arr>4] #4'den büyük olanları getirir
```

```python
arr[(arr>3)&(arr<8)] #3'den büyük ve 8'den küçük olanları getirir
```

```python
arr[arr!=4] #4'e eşit olmayanları getirir
```

```python
arr[arr<=6]
```

# Broadcasting

Numpy arrays differ from a normal Python list because of their ability to broadcast:

https://numpy.org/doc/stable/user/theory.broadcasting.html

```python
arr
```

```python
arr[:5] = 50 #5'e kadar olanların hepsini 50 yapar
arr
```

```python
arr
```

```python
arr_slice = arr[4:] #arr içinde belirli bir parça almak için. 4. row dan itibaren hepsini al
arr_slice
```

```python
arr_slice[-2:] = 55 #sondan 2 indeksi 55 yapar
arr_slice
```

## Statistical Calculations

* ``np.mean(arr,axis=0)`` | Returns mean along specific axis

* ``arr.sum()`` | Returns sum of arr

* ``arr.min()`` | Returns minimum value of arr

* ``arr.max(axis=0)`` | Returns maximum value of specific axis

* ``np.var(arr)`` | Returns the variance of array

* ``np.std(arr,axis=1)`` | Returns the standard deviation of specific axis

* ``np.corrcoef(arr)`` | Returns correlation coefficient of array

```python
v=np.array([1,1,2,2,3,3,3])
v
```

```python
np.mean(v) #v.mean() da aynı iişlemi yapar
```

```python
np.median(v)
```

```python
np.std(v)
```

```python
np.var(v)
```

## Excel
Pandas can read and write excel files, keep in mind, this only imports data. Not formulas or images, having images or macros may cause this read_excel method to crash. 

### Excel Input

```python
df1=pd.read_excel("Excel_Sample.xlsx") #Bu otomatik olarak sheet1 i okudu. Çünkü excel içinde iki tane sheet var.
df1
```

```python
df2=pd.read_excel("Excel_Sample.xlsx", sheet_name="Sheet2") #Sheet2 yi okudu
df2
```

```python
df3=pd.read_excel("Excel_Sample.xlsx", sheet_name=None) #None yazarsak da Sheet lerin hepsine bakabiliriz.
df3
```

```python
type(df3)
```

```python
pd.ExcelFile("Excel_Sample.xlsx").sheet_names #Bu da içerisinde bulunan Sheet isimlerini verir.
```

### Excel Output

```python
df1.to_excel("Excel_Sample1.xlsx", sheet_name="Sheet1", index=False)
```

```python
df2.to_excel("Excel_Sample1.xlsx", sheet_name="Sheet2", index=False) #Ayrı bir excel yaptı ayrı ayrı sheet yapmadı
```

```python
with pd.ExcelWriter('combined_df.xlsx') as writer:    #Bu şekilde yaparsak aynı excel içinde iki farklı sheet yapabiliriz
    df1.to_excel(writer, sheet_name='sample1', index=False)
    df2.to_excel(writer, sheet_name='sample2', index=True)
```

```python
pd.read_excel("combined_df.xlsx", sheet_name="sample1")
```

```python
pd.read_excel("combined_df.xlsx", sheet_name="sample2")
```

# SQL Connections


```python
pip install sqlalchemy
```

```python
from sqlalchemy import create_engine #Python ile SQL bağlantısı kuracak makina
```

```python
temp_db = create_engine("sqlite:///:memory") #Kendi bilgisayarımızın ram inde irtibatı sağlayacak dataframe oluşturur. Geçici olarak oluşur bu.
```

```python
temp_db.table_names() #Boş bir database oluştu şu an
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
```

```python
tables
```

```python
tables[6]
```

```python
tables[6].to_sql(name="populations", con=temp_db) #table 6 yı populasyon adında temp_db connection ile at.
```

```python
temp_db.table_names() #İçerisinde olduğunu artık görebiliriz.
```

```python
pd.read_sql(sql="populations", con=temp_db)
```

```python
pd.read_sql_query(sql="SELECT Country, Rank FROM populations", con=temp_db)
```

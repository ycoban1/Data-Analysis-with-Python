## Multi-Index and Index Hierarchy

Let us go over how to work with Multi-Index, first we'll create a quick example of what a Multi-Indexed DataFrame would look like:

```python
outside = ['M1', 'M1', 'M1', 'M2', 'M2', 'M2','M3', 'M3', 'M3']
inside = [1, 2, 3, 1, 2, 3, 5, 6, 7]
multi_index = list(zip(outside, inside))
multi_index
```

```python
hier_index=pd.MultiIndex.from_tuples(multi_index)
```

```python

df=pd.DataFrame(np.random.randn(9,4), index = hier_index, columns=['A','B','C','D'])
df
```

```python
df.loc["M1"]
```

```python
df.loc["M1"].loc[2]
```

```python
df.loc["M1"].loc[[2]] #DataFrame olarak verdi
```

```python
df.index.names #index isimlerini verir
```

```python
df.index.names = ["Group" , "Num"] #index leri isimlendirdik
```

```python
df
```

```python
df.xs("M1")
```

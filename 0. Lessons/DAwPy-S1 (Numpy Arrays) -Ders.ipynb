# NumPy 

NumPy (or Numpy) is a Linear Algebra Library for Python, the reason it is so important for Data Science with Python is that almost all of the libraries in the PyData Ecosystem rely on NumPy as one of their main building blocks.

Numpy is also incredibly fast, as it has bindings to C libraries. For more info on why you would want to use Arrays instead of lists, check out this great [StackOverflow post](http://stackoverflow.com/questions/993984/why-numpy-instead-of-python-lists).

We will only learn the basics of NumPy, to get started we need to install it!

## Installation Instructions

**It is highly recommended you install Python using the Anaconda distribution to make sure all underlying dependencies (such as Linear Algebra libraries) all sync up with the use of a conda install. If you have Anaconda, install NumPy by going to your terminal or command prompt and typing:**
    
    conda install numpy
    
**If you do not have Anaconda and can not install it, please refer to [Numpy's official documentation on various installation instructions.](http://docs.scipy.org/doc/numpy-1.10.1/user/install.html)**

## Using NumPy

Once you've installed NumPy you can import it as a library:

```python
import numpy as np
```

Numpy has many built-in functions and capabilities. We won't cover them all but instead we will focus on some of the most important aspects of Numpy: vectors,arrays,matrices, and number generation. Let's start by discussing arrays.

# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

### Built-in Methods

There are lots of built-in ways to generate Arrays

### ``arange``

Return evenly spaced values within a given interval.

arange([start,] stop[, step,], dtype=None)

```python
np.arange(10) #0'dan 10 (hariç)'a kadar sıralar.
```

```python
np.arange(2,10) #2'dan 10 (hariç)'a kadar sıralar.
```

```python
np.arange(0,10,2) #0'dan 10 (hariç)'a kadar 2 şer sıralar.
```

### ``linspace``
Return evenly spaced numbers over a specified interval.

np.linspace(start, stop, num=50)

```python
np.linspace(0,10,3) #0 dan 10 a 3 sayı ver
```

```python
np.arange(0,11,5) #np.linspace(0,10,3) ikisi aynı iş oluyor
```

```python
np.linspace(0,10,50)
```

```python
np.linspace(0,10)
```

```python
np.linspace(0,10, dtype=int)
```

```python
np.linspace(0,10, dtype=float)
```

Generate arrays of zeros or ones

np.zeros(shape, dtype=float, order='C')

np.ones(shape, dtype=None, order='C')

np.full(shape, fill_value, dtype=None, order='C')

```python
np.zeros(10, dtype=str)
```

```python
np.ones(5)
```

```python
np.full((3,4),5) #3x4 lük 5 lerden oluşan array
```

### Random 

Numpy also has lots of ways to create random number arrays:

### ``rand``
Create an array of the given shape and populate it with
random samples from a uniform distribution
over ``[0, 1)``.

- ``(0, 5)`` = 1, 2, 3, 4
- ``(0, 5]`` = 1, 2, 3, 4, 5
- ``[0, 5)`` = 0, 1, 2, 3, 4
- ``[0, 5]`` = 0, 1, 2, 3, 4, 5

```python
np.random.rand(2) #0 ile 1 arasında (1 dahil değil) 2 tane random sayı üretim.
```

```python
np.random.rand(5,4) #0 ile 1 arasında (1 dahil değil) 5x4 random sayı üretim.
```

```python
plt.hist(np.random.rand(555), bins= 40) #555 tane 40 aralıklarla random
plt.show();
```

### ``randn``

Return a sample (or samples) from the "standard normal" distribution. Unlike rand which is uniform:

```python
np.random.randn(2) #mean ı olan standart sapması 1 olan random 2 sayı
```

```python
np.random.randn(5,5)
```

```python
plt.hist(np.random.randn(555), bins= 40) #555 tane 40 aralıklarla random
plt.show();
```

### ``randint``
Return random integers from `low` (inclusive) to `high` (exclusive).

randint(low, high=None, size=None, dtype=int)

```python
np.random.randint(10) #0-10 arasında random 1 sayı
```

```python
np.random.randint(1,10) #1 (dahil) ile 10 (hariç) arasında random bir sayı
```

```python
np.random.randint(100, size=(5)) #0 ile 100 (hariç) arasında sayılardan oluşan 5 random sayı
```

```python
np.random.randint(100, size=(2,5)) #0 ile 100 (hariç) arasında random sayılardan oluşan 2x5 
```

```python
np.random.randint(1,[3,5,50], size=(5,3)) #1-3 arasında, 1-5 arasında, 1-50 arasında üç 5x3
```

# Concatenation of the Arrays

concatenate((a1, a2, ...), axis=0, out=None)

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
z = np.array([7, 8, 9])
```

```python
np.concatenate([x,y,z]) #üç array ı birleştirir
```

```python
x.reshape(1,3) #x in boyununu artırır
```

```python
np.concatenate([x.reshape(1,3), y.reshape(1,3), z.reshape(1,3)], axis = 0) #önce hepsinin boyununu artırdık
```

```python
a = np.random.randint(10, size = (2,3))
b = np.random.randint(10, size = (2,3))
a
```

```python
np.concatenate([a,b])
```

```python
np.concatenate([a,b], axis=1)
```

# Splitting of the Arrays

np.split(ary, indices_or_sections, axis=0)

```python
x = np.array([1, 2, 3, 99, 99, 3, 2, 1])
x
```

```python
np.split(x,[3,5]) #x i önce 3. indekse kadar böl ve sonra 5. indekse kadar daha böl. Sonra kalanı böl.
```

```python
np.split(x,[4])
```

```python
np.split(x,4) #x i eşit olarak 4 e böler
```
